```python
import pandas as pd
import numpy as nm
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/bpp.csv') #reading the dataset into dataframe hello
```

```python
df.describe().T
```

```python
X = df.drop(["SalePrice","ModelID","MachineID","SalesID","saleMonth","auctioneerID"],axis=1) #Input features (remove features with null values,unwanted metrics and un processed data)
y= df[["SalePrice"]] #Output
```

```python
df=df.dropna(axis=0)
```

```python
from sklearn.model_selection import train_test_split #splitting the data test size 25%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler # scaling the data or normalizing the data z = (x - u) / s (u=mean and s=standard deviation)
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train) # scaling both train data and test data
X_test_scaled = ss.transform(X_test)
```

```python
from sklearn.linear_model import Lasso  # Lasso regression model

lasso_model = Lasso(alpha=1.0)  # Alpha is the regularization strength
lasso_model.fit(X_train_scaled, y_train)

lasso_model.intercept_  # the constant value
```

```python
lasso_model.coef_  # the weights
```

```python
y_pred = lasso_model.predict(X_test_scaled)  # prediction
```

```python
result = pd.DataFrame()
result["Predictions_LR"] = nm.reshape(y_pred,75000)
result["Actual"] = y_test.values
```

```python
result # result data frame
```

```python
from sklearn.metrics import mean_absolute_percentage_error #mean of (yactual-ypred/yactual)
mean_absolute_percentage_error(result["Actual"],result["Predictions_LR"])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Assuming you already have X_train_scaled, y_train, X_test_scaled, y_test

# Create a Lasso model
lasso_model = Lasso(max_iter=1000)

# Define a range of alpha values to search over
alphas = [0.01, 0.1, 1.0, 10.0]

# Perform a grid search using cross-validation
param_grid = {'alpha': alphas}
grid_search = GridSearchCV(lasso_model, param_grid, cv=5, scoring='neg_mean_absolute_error')
grid_search.fit(X_train_scaled, y_train)

# Get the best alpha from the grid search
best_alpha = grid_search.best_params_['alpha']
print("Best alpha:", best_alpha)

# Create a new Lasso model with the best alpha and fit it to the data
best_lasso_model = Lasso(alpha=best_alpha)
best_lasso_model.fit(X_train_scaled, y_train)

# Make predictions using the best model
y_pred = best_lasso_model.predict(X_test_scaled)

# Evaluate the model's performance
from sklearn.metrics import mean_absolute_percentage_error, r2_score

mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Absolute Percentage Error:", mape)
print("R-squared:", r2)
```
